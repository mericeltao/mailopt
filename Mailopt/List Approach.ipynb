```python
"""
Most recent version of m/m/1 queue simulation
*takes several seconds to compile (output is at bottom of cell)

#average waiting times should be lambda/mu something 

"""
import cvxpy
import numpy as np
import random as rm
import random
from scipy.stats import poisson
from scipy.stats import expon
from re import X

class MailroomSim:
    #m/m/1 queue with exponential distribution (mean 1) for arrival and service times
    def __init__(self):
      """
      Initialize class instances below
      
      arrival_times_list: randomly generated list of all arrival times of students
      departure_times_list: calculated departure times based on arrival times, employee availability, and randomly generated service times
      waitTimes: list of wait times which corresponds to each arrival 
      """

      self.arrival_times_list = []
      self.newDeparturesList = []
      self.waitTimes = []
      
      self.lamb = 0.5
      self.mu = 1

      
      #lamb and mu were sayingt they weren't defined, i think if you initiaize
      #self.lamb and self.mu instead, it should work

    def increment_time(self): 
      """
      Creates output of all arrival and departure times within 8am-10pm Monday-Friday business hours
      Generates arrival times in an exponential distrbution 
      """
      count_time = 0.0

      #random_exp_list = expon.rvs(scale = self.lamb, size = 10000)
      while count_time <= 10000:
        #count_time += np.random.choice(random_exp_list)
        count_time += random.expovariate(self.lamb)
        #For all values in the randomly generated list, adds the values to count time
        if count_time <= 840:
          self.arrival_times_list.append(count_time)    
          #print(count_time)
        
      currentClock = self.arrival_times_list[0]+ self.service_time()
      lastDeparture = currentClock
      self.newDeparturesList.append(lastDeparture)

      for x in self.arrival_times_list[1:]:
        if x > lastDeparture:
          currentClock = x
          lastDeparture = currentClock + self.service_time()
          self.newDeparturesList.append(lastDeparture)
          currentClock = lastDeparture
        else: 
          currentCock = lastDeparture
          lastDeparture = currentClock + self.service_time()
          self.newDeparturesList.append(lastDeparture)
          currentClock = lastDeparture

      sum = 0
      count = 0
      for i in range(len(self.newDeparturesList)):
        waitTime = self.newDeparturesList[i] - self.arrival_times_list[i]
        sum+= waitTime
        count+=1
        self.waitTimes.append(waitTime)
      
      average = sum/count

      print("Arrivals:")
      print(self.arrival_times_list)
      print("Departures:")
      print(self.newDeparturesList)
      print("Wait Times:")
      print(self.waitTimes)
      print("average wait time:")
      print(average)


    #how long something takes to process depending on mail or other request (values can be modified based on observed real data)
    def service_time(self):
      service_time = np.random.exponential(1/self.mu)
      return service_time
      
s = MailroomSim()
s.increment_time()
```
